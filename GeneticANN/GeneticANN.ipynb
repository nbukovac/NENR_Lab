## Zadatak 1

```python
import matplotlib.pyplot as plt
import numpy as np

f = lambda x, w, s : 1.0 / (1 + abs(x - w) / abs(s))
x = np.linspace(-8.0, 10.0)
y1 = []
y2 = []
y3 = []

w0 = 2
s1 = 1
s2 = 0.25
s3 = 4

for i in x:
    y1.append(f(i, w0, s1))
    y2.append(f(i, w0, s2))
    y3.append(f(i, w0, s3))
              
plt.figure(figsize=(16.,10.))
        
plt.plot(x, y1, c='red', label='s = 1')
plt.plot(x, y2, c='blue', label='s = 0.25')
plt.plot(x, y3, c='black', label='s = 4')

plt.legend()
plt.show()
```

## Zadatak 2

```python
x = []
y = []
classified = []

with open('dataset.txt') as f: 
    for line in f:
        split = line.split()
        x.append(float(split[0]))
        y.append(float(split[1]))
        
        if split[2] == '1':
            classified.append('red')
        elif split[3] == '1':
            classified.append('blue')
        else:
            classified.append('black')
            
plt.figure(figsize=(16.,10.))
plt.scatter(x, y, c=classified)
plt.show()
```
